# Crivo de Erastótenes

```python
def Sieve_E(n):
    
    # Crivo de Erastótenes
    A = [True]*(n+1)
    for i in range(2, int(np.ceil(np.sqrt(n))) ):
        if A[i] == True:
            j = i**2
            while j<=n:
                A[j]=False
                j += i
                
    primes = [i for i, x in enumerate(A) if x == True]
    primes.pop(0) # removo 0
    primes.pop(0) # removo 1
    
    return primes
```
